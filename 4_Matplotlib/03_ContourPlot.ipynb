# Contour Plots

````python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(0)
````

````python
def f(x, y):
    return x**2 + y**2
````

````python
x = np.arange(-5, 5.0, 0.25)
y = np.arange(-5, 5.0, 0.25)
````

````python
print(x[:10])
print(y[:10])
````

### Meshgrid


Return coordinate matrices from coordinate vectors.

````python
X, Y = np.meshgrid(x, y)

print(X)
print(Y)
````

````python
plt.scatter(X, Y, s=10);
````

````python
Z = f(X, Y)

print(Z)
````

````python
plt.contour(X, Y, Z, colors='black');
````

````python
delta = 0.025

x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)

X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
````

````python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
````
