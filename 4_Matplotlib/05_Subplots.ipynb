# Multiple Subplots

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(0)
```

### add_subplot

The calling signature of add_subplot does not directly provide the option to place the axes at a predefined position. It rather allows to specify where the axes should be situated according to a subplot grid. The usual and easiest way to specify this position is the 3 integer notation.

```python
fig = plt.figure()

ax1 = fig.add_subplot(211)
ax1.set_ylim((-1.2, 1.2))

ax2 = fig.add_subplot(212)
ax2.set_ylim((-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

## Grids of Subplots

```python
for i in range(1, 3):
    plt.subplot(2, 1, i)
    if i == 1:
        plt.plot(np.sin(x))
    else:
        plt.plot(np.cos(x))
```

```python
ax1 = plt.subplot(2, 1, 1)
ax1.plot(np.sin(x))
ax2 = plt.subplot(2, 1, 2)
ax2.plot(np.cos(x));
```

```python
fig, ax = plt.subplots(2, 1, sharex='col', sharey='row')

print(type(ax))

ax[0].plot(np.sin(x))
ax[1].plot(np.cos(x));
```

```python
fig, ax = plt.subplots(2, 2, sharex='col', sharey='row')

print(type(ax))

for i in range(2):
    for j in range(2):
        ax[i, j].plot(np.sin(x))

fig;
```

## Additional Info

`fig.add_subplot` can also handle multiple rows and cols

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
ax1.set_ylim((-1.2, 1.2))

ax2 = fig.add_subplot(222)
ax2.set_ylim((-1.2, 1.2))

ax3 = fig.add_subplot(223) # 3 = 2nd row, 1st col
ax3.set_ylim((-1.2, 1.2))

ax4 = fig.add_subplot(224) # 3 = 2nd row, 2nd col
ax4.set_ylim((-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
ax3.plot(np.cos(x))
ax4.plot(np.cos(x));
```
